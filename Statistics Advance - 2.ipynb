```python
"""
1. What is hypothesis testing in statistics?

-> Hypothesis testing is a statistical method used to decide if there is enough evidence 
from sample data to support a specific claim about a population. 

It involves:
a. Defining a null hypothesis, which is the default assumption(for example, no effect or no difference).
b. Defining an alternative hypothesis, which is the claim we want to test.
c. Calculating a test statistic and a p-value from the data.
d. Comparing the p-value to a chosen significance level, usually 0.05.
e. Rejecting the null hypothesis if the p-value is less than or equal to the significance level,
meaning there is evidence for the alternative hypothesis; otherwise, failing to reject the null hypothesis.

It helps make informed decisions based on data.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

-> The null hypothesis is the default assumption in hypothesis testing. It usually states that there is no effect, 
no difference, or no relationship between variables. For example, a new drug has no impact on patient recovery.

The alternative hypothesis is what we want to test or prove. It suggests that there is an effect, 
difference, or relationship. For example, "the new drug improves patient recovery."

Difference:

a. The null hypothesis assumes no change or status quo.
b. The alternative hypothesis proposes a change or effect.

In testing, we try to find evidence to reject the null hypothesis in favor of the alternative.

3. What is the significance level in hypothesis testing, and why is it important?

-> The significance level in hypothesis testing is a threshold (commonly set at 0.05) 
that defines how much risk of making a wrong decision we are willing to accept. 
Specifically, it is the probability of rejecting the null hypothesis when it is actually true (a false positive).

It’s important because:

a. It controls the chance of making this error, called a Type I error.
b. It helps decide when the evidence against the null hypothesis is strong enough to reject it.
c. Setting it beforehand ensures objective and consistent decision-making in tests.

4. What does a P-value represent in hypothesis testing?

-> A p-value represents the probability of observing the data—or something 
more extreme—assuming the null hypothesis is true.

It tells us how likely our results would occur by chance if there was actually no effect.

A small p-value (usually less than 0.05) suggests that the observed data is unlikely under the null hypothesis, 
providing evidence to reject it. A large p-value means there isn’t enough evidence to reject the null hypothesis.

5. How do you interpret the P-value in hypothesis testing?

-> To interpret the p-value in hypothesis testing:

a. If the p-value is less than or equal to the significance level (usually 0.05):
-> Reject the null hypothesis.
-> There is enough evidence to support the alternative hypothesis.

b. If the p-value is greater than the significance level:
-> Fail to reject the null hypothesis.
-> There is not enough evidence to support the alternative hypothesis.

A low p-value means the result is statistically significant; a high p-value means it is not.

6. What are Type 1 and Type 2 errors in hypothesis testing?

-> In hypothesis testing, Type 1 and Type 2 errors are possible mistakes in decision-making:

a. Type 1 Error: Rejecting the null hypothesis when it is actually true.
-> It's a false positive.
-> The probability of this error is the significance level (usually 0.05).

b. Type 2 Error: Failing to reject the null hypothesis when the alternative hypothesis is true.
-> It's a false negative.
-> The probability of this error is called beta, and (1-beta) is the power of the test.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

-> The difference between a one-tailed and a two-tailed test in hypothesis testing lies 
in the direction of the effect being tested:

a. One-tailed test:
-> Tests for an effect in only one direction (either greater than or less than).
-> Example: Testing if a drug increases recovery rate (only checking for improvement).
-> More powerful for detecting an effect in the specified direction.
-> Used when we have a clear expectation of the effect’s direction.

b. Two-tailed test:
-> Tests for an effect in both directions (either greater or less).
-> Example: Testing if a new teaching method has a different effect (better or worse) than the standard.
-> More conservative, but safer when we’re unsure of the direction.

8. What is the Z-test, and when is it used in hypothesis testing?

-> The Z-test is a statistical test used to determine whether there is a significant difference 
between sample data and a population, or between two samples, 
when the population standard deviation is known and the sample size is large (typically 30 or more).

Z-test is used when:
a. The population variance or standard deviation is known.
b. The sample size is large (Central Limit Theorem applies).
c. The data is approximately normally distributed.

Example use:
Testing if the average test score of a class differs from the national average, 
assuming the population standard deviation is known.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing?

-> To calculate the Z-score, subtract the population mean from the data point or sample mean, 
then divide the result by the population standard deviation.

If we are working with a sample mean, divide the population standard deviation by the square root of the sample size 
(this is called the standard error) before using it in the calculation.

The Z-score tells us how many standard deviations a value is away from the population mean. 
It helps measure how unusual or extreme a value is, assuming the null hypothesis is true.

Interpretation:
a. A Z-score close to zero means the value is near the average.
b. A large positive or negative Z-score means the value is far from the average.

In hypothesis testing, a high absolute Z-score suggests the result is statistically significant 
and likely not due to random chance.

10. What is the T-distribution, and when should it be used instead of the normal distribution?

-> The T-distribution is a probability distribution used in statistics 
when estimating population parameters when the sample size is small and the population standard deviation is unknown.

When to use the T-distribution instead of the normal distribution:
a. The sample size is small(typically less than 30).
b. The population standard deviation is unknown.
c. The data is approximately normally distributed.

It's important because:
a. The T-distribution has heavier tails than the normal distribution, 
which gives more room for uncertainty when dealing with small samples.
b. As the sample size increases, the T-distribution becomes more similar to the normal distribution.

Example:
If we're testing whether the average height of a small group differs from a known average 
and we don’t know the population standard deviation, we should use the T-distribution.

11. What is the difference between a Z-test and a T-test?

->  Z-test:
a. Used when the population standard deviation is known.
b. Sample size is large(usually 30 or more).
c. Assumes data follows a normal distribution.
d. More precise when population parameters are known.

T-test:
a. Used when the population standard deviation is unknown.
b. Sample size is small (typically less than 30).
c. Uses the sample standard deviation instead of the population's.
d. Based on the T-distribution, which adjusts for extra uncertainty in small samples.

12. What is the T-test, and how is it used in hypothesis testing?

-> The T-test is a statistical test used to determine whether there is a significant difference between means. 
It is especially useful when the sample size is small and the population standard deviation is unknown.

It's used in hypothesis testing:

a. State hypotheses:

-> Null hypothesis: No difference in means
-> Alternative hypothesis: There is a difference in means

b. Calculate the T-statistic:

Measures how far the sample mean is from the population mean (or between sample means), adjusted for sample variability.

c. Find the P-value:

Shows the probability of observing the result if the null hypothesis is true.

d. Compare P-value to significance level (like 0.05):

-> If P-value is less than the significance level, reject the null hypothesis.
-> Otherwise, fail to reject the null hypothesis.

Example:
Testing whether a new drug changes blood pressure. We compare the mean blood pressure of a sample 
who took the drug to a known average using a one-sample T-test.

13. What is the relationship between Z-test and T-test in hypothesis testing?

-> The Z-test and T-test are closely related—they both test hypotheses about population means, 
but they differ in when and how they are used.

Relationship between Z-test and T-test in hypothesis testing:

a. Both tests compare sample data to population parameters to determine 
if observed differences are statistically significant.

b. The main difference lies in what is known about the population and the sample size.

Both tests serve the same purpose—to evaluate if a difference is statistically significant—
but the T-test is more flexible and preferred when the population standard deviation is unknown or the sample size is small.

14. What is a confidence interval, and how is it used to interpret statistical results?

-> A confidence interval is a range of values, calculated from sample data, 
that is likely to contain the true population parameter (such as the mean) with a certain level of confidence (often 95%).

It is used to interpret statistical results:

a. It provides an estimate of uncertainty around the sample statistic.
For example, a 95% confidence interval means that if we repeated the study many times, 
about 95% of the calculated intervals would contain the true population value.

b. It helps us understand how precise our estimate is.

c. If the confidence interval is narrow, it means our estimate is more precise.

d. If it is wide, there is more uncertainty in the estimate.

e. Confidence intervals are also used to test hypotheses:
If a hypothesized value (like a population mean) falls outside the confidence interval, 
it suggests that the hypothesis may be unlikely or rejected.

15. What is the margin of error, and how does it affect the confidence interval?

-> The margin of error is the amount added and subtracted from a sample estimate (like a mean) 
to create the confidence interval. It represents the maximum expected difference 
between the true population parameter and the sample estimate due to random sampling.

It affects the confidence interval:
a. The margin of error determines the width of the confidence interval.
b. A larger margin of error means a wider confidence interval, indicating more uncertainty.
c. A smaller margin of error means a narrower confidence interval, indicating more precise estimates.
d. It depends on factors like the sample size, variability in the data, and the confidence level chosen.

16. How is Bayes' Theorem used in statistics, and what is its significance?

-> Bayes' Theorem provides a way to update the probability of a hypothesis based on new evidence or data. 
It combines prior knowledge (prior probability) with new data (likelihood) to calculate an updated probability 
(posterior probability).
Formula (conceptually): Posterior Probability = (Likelihood × Prior Probability) / Evidence

Significance of Bayes' Theorem:
a. Incorporates prior knowledge: It allows statisticians to formally include previous information or beliefs.
b. Updates beliefs: As new data arrives, it refines the probability of an event or hypothesis.
c. Foundation of Bayesian statistics: Used widely in machine learning, decision making, and many fields where uncertainty is involved.
d. Helps in making decisions under uncertainty and improving predictions.

Example:
If we want to know the probability a patient has a disease after a positive test, 
Bayes’ Theorem combines the test accuracy and the prior likelihood of the disease to give a more accurate probability.

17. What is the Chi-square distribution, and when is it used?

-> The Chi-square distribution is a probability distribution that describes the sum of the squares of 
independent standard normal random variables. It is asymmetric and skewed right, with its shape 
depending on the degrees of freedom (which relates to sample size).

It is used when:
a. Goodness-of-fit tests: To check how well observed data fits an expected distribution.
b. Test for independence: In contingency tables to test if two categorical variables are related.
c. Test for homogeneity: To compare distributions across different populations.
d. Variance estimation: Used in confidence intervals and hypothesis tests involving population variances.

18. What is the Chi-square goodness of fit test, and how is it applied?

-> The Chi-square goodness of fit test is a statistical test used to determine whether observed categorical data 
fits an expected distribution. It checks if the differences between observed and expected frequencies 
are due to chance or if they are statistically significant.

It is applied:

a. Define hypotheses:
-> Null hypothesis: The observed data fits the expected distribution.
-> Alternative hypothesis: The observed data does not fit the expected distribution.

b. Collect data:
Obtain observed frequencies for each category.

c. Determine expected frequencies:
Based on the expected distribution, calculate the expected counts for each category.

d. Calculate the Chi-square statistic:
Sum over all categories the squared difference between observed and expected counts divided by the expected counts.

e. Compare statistic to critical value:
Using Chi-square distribution with appropriate degrees of freedom, find the critical value or P-value.

f. Make decision:

-> If the Chi-square statistic is greater than the critical value or P-value is less than significance level, 
reject the null hypothesis.
-> Otherwise, fail to reject the null hypothesis.

19. What is the F-distribution, and when is it used in hypothesis testing?

-> The F-distribution is a probability distribution that arises when comparing 
the ratio of two sample variances.. It is right-skewed and depends on two sets of degrees of freedom 
(numerator and denominator).

It is used in hypothesis testing:

a. Primarily used in ANOVA (Analysis of Variance) to compare the variances of three or more groups 
to see if their means differ significantly.
b. Used to test the equality of variances between two populations (e.g., in the F-test for variance).
c. Applied in regression analysis to test the overall significance of a model.

20. What is an ANOVA test, and what are its assumptions?

-> ANOVA (Analysis of Variance) is a statistical method used to compare the means of three or more groups 
to see if at least one group mean is significantly different from the others. It tests whether observed differences 
among group means are likely due to random chance or a real effect.

Assumptions of ANOVA:
a. Independence: Observations in each group are independent of each other.
b. Normality: The data in each group should be approximately normally distributed.
c. Homogeneity of variances (Equal variances): The variances across the groups should be roughly equal.
d. Absence of outliers: There should be no significant outliers in the data, as they can distort the results.

21. What are the different types of ANOVA tests?

-> The main types of ANOVA tests are:

a. One-Way ANOVA:
Compares the means of three or more independent groups based on one factor or independent variable.

b. Two-Way ANOVA:
Examines the effect of two independent factors on a dependent variable, including the interaction between the factors.

c. Repeated Measures ANOVA:
Used when the same subjects are measured multiple times under different conditions or over time.

d. Multivariate ANOVA (MANOVA):
Extends ANOVA to analyze multiple dependent variables simultaneously.

22. What is the F-test, and how does it relate to hypothesis testing?

-> The F-test is a statistical test that compares variances to determine if they are significantly different. 
It uses the F-distribution and is often applied in the context of ANOVA, regression, 
or testing the equality of two variances.

It relates to hypothesis testing:
a. The F-test helps test null hypotheses about the equality of variances or model fit.
b. In ANOVA, the F-test compares between-group variance to within-group variance.
c. In regression, it checks if the overall regression model is significant 
(i.e., if the model explains a significant portion of the variance in the response variable).
"""
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(x, p):
    if x not in [0,1]:
        raise ValueError("x must be 0 or 1 for Bernoulli distribution.")
    if not (0<=p<=1):
        raise ValueError("Probability p must be between 0 and 1.")
    return p**x*(1-p)**(1-x)
p=0.6
print(bernoulli_pdf(1,p))
print(bernoulli_pdf(0,p))
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
def calculate_and_plot_z_scores(data):
    data=np.array(data)
    mean=np.mean(data)
    std=np.std(data)
    if std==0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be computed.")
    z_scores=(data-mean)/std
    plt.figure(figsize=(8,5))
    plt.hist(z_scores,bins=30,density=True,edgecolor='black',alpha=0.7)
    plt.title("Z-score Distribution")
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
    return z_scores
sample_data=np.random.normal(100, 15, 1000)
z_scores=calculate_and_plot_z_scores(sample_data)
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm
mu=60000   
sample_mean=58000   
sigma=5000    
n=50     
alpha=0.05
z_stat=(sample_mean-mu)/(sigma/np.sqrt(n))
z_critical=norm.ppf(1-alpha/2)
p_value=2*(1-norm.cdf(abs(z_stat)))
print(f"Z-statistic: {z_stat:.2f}")
print(f"Z-critical (±): {z_critical:.2f}")
print(f"P-value: {p_value:.4f}")
if abs(z_stat)>z_critical:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
```

```python
# 16. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats
data=[52,56,59,63,61,58,62,60,57,55]
sample_mean=np.mean(data)
sample_std=np.std(data,ddof=1)
n=len(data)
confidence=0.95
alpha=1-confidence
t_critical=stats.t.ppf(1-alpha/2,df=n-1) 
margin_of_error=t_critical*(sample_std/np.sqrt(n))
lower_bound=sample_mean-margin_of_error
upper_bound=sample_mean+margin_of_error
print(f"Sample Mean: {sample_mean:.2f}")
print(f"{confidence*100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
# 1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm
def z_test(sample,pop_mean,pop_std,alpha=0.05):
    n=len(sample)
    sample_mean=np.mean(sample)
    z_stat=(sample_mean-pop_mean)/(pop_std/np.sqrt(n))
    p_value=2*(1-norm.cdf(abs(z_stat)))
    print(f"Sample Mean = {sample_mean:.4f}")
    print(f"Population Mean = {pop_mean}")
    print(f"Z statistic = {z_stat:.4f}")
    print(f"P-value = {p_value:.4f}")
    if p_value<alpha:
        print(f"Reject the null hypothesis at α = {alpha}")
        print("Conclusion: The sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}")
        print("Conclusion: No significant difference between sample mean and population mean.")
np.random.seed(42)
sample_data=np.random.normal(loc=52,scale=10,size=30)
population_mean=50
population_std=10
z_test(sample_data,population_mean,population_std)
```

```python
# 3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy.stats import norm
def one_sample_z_test(sample,pop_mean,pop_std,alpha=0.05):
    n=len(sample)
    sample_mean=np.mean(sample)
    z_stat=(sample_mean-pop_mean)/(pop_std/np.sqrt(n))
    p_value=2*(1-norm.cdf(abs(z_stat)))
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Population Mean: {pop_mean}")
    print(f"Z-Statistic: {z_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    if p_value < alpha:
        print(f"Reject the null hypothesis at alpha = {alpha}")
        print("Conclusion: Sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at alpha = {alpha}")
        print("Conclusion: No significant difference between sample mean and population mean.")
np.random.seed(123)
sample_data=np.random.normal(loc=52,scale=10,size=40)
population_mean=50
population_std=10
one_sample_z_test(sample_data,population_mean,population_std)
```

```python
# 4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def two_tailed_z_test(sample,pop_mean,pop_std,alpha=0.05):
    n=len(sample)
    sample_mean=np.mean(sample)
    z_stat=(sample_mean-pop_mean)/(pop_std/np.sqrt(n))
    p_value=2*(1-norm.cdf(abs(z_stat)))
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Population Mean: {pop_mean}")
    print(f"Z-Statistic: {z_stat:.4f}")
    print(f"P-Value: {p_value:.4f}")
    z_crit =norm.ppf(1-alpha/2)
    print(f"Critical Z-Values: ±{z_crit:.4f}")
    if abs(z_stat)>z_crit:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")
    x=np.linspace(-4,4,1000)
    y=norm.pdf(x)
    plt.figure(figsize=(10,6))
    plt.plot(x,y,label='Standard Normal Distribution')
    plt.fill_between(x,0,y,where=(x<=-z_crit),color='red',alpha=0.5,label='Rejection Region (Left)')
    plt.fill_between(x,0,y,where=(x>=z_crit),color='red',alpha=0.5,label='Rejection Region (Right)')
    plt.axvline(z_stat,color='blue',linestyle='--',linewidth=2,label=f'Z Statistic={z_stat:.2f}')
    plt.title('Two-Tailed Z-Test Decision Regions')
    plt.xlabel('Z value')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
np.random.seed(101)
sample=np.random.normal(loc=52,scale=10,size=40) 
population_mean=50
population_std=10
alpha=0.05
two_tailed_z_test(sample,population_mean,population_std,alpha)
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def plot_type1_type2(pop_mean,pop_std,n,alpha,alt_mean):
    se=pop_std/np.sqrt(n)
    z_crit=norm.ppf(1-alpha/2)
    crit_low=pop_mean-z_crit*se
    crit_high=pop_mean+z_crit*se
    beta_low=norm.cdf(crit_low,loc=alt_mean,scale=se)
    beta_high=1-norm.cdf(crit_high,loc=alt_mean,scale=se)
    beta=beta_high+beta_low
    power=1-beta
    print(f"Type 1 error: {alpha}")
    print(f"Type 2 error: {beta:.4f}")
    print(f"Power: {power:.4f}")
    x=np.linspace(pop_mean-4*se,alt_mean+4*se,1000)
    pdf_h0=norm.pdf(x,loc=pop_mean,scale=se)
    pdf_h1=norm.pdf(x,loc=alt_mean,scale=se)
    plt.figure(figsize=(12,6))
    plt.plot(x,pdf_h0,label='Null Hypothesis (H0)',color='blue')
    plt.plot(x,pdf_h1,label='Alternative Hypothesis (H1)',color='green')
    plt.fill_between(x,0,pdf_h0,where=(x<=crit_low),color='red',alpha=0.3,label='Rejection Region')
    plt.fill_between(x,0,pdf_h0,where=(x>=crit_high),color='red',alpha=0.3)
    plt.fill_between(x,0,pdf_h1,where=((x>crit_low)&(x<crit_high)),color='orange',alpha=0.3,label='Type 2 Error')
    plt.axvline(crit_low,color='red',linestyle='--')
    plt.axvline(crit_high,color='red',linestyle='--')
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()
pop_mean=50        
pop_std=10         
sample_size=30      
alpha=0.05          
alt_mean=54                                                                       
plot_type1_type2(pop_mean,pop_std,sample_size,alpha,alt_mean)
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind
def independent_t_test(sample1,sample2,alpha=0.05):
    t_stat,p_value=ttest_ind(sample1,sample2,equal_var=False)
    print("Sample 1 Mean:",np.mean(sample1))
    print("Sample 2 Mean:",np.mean(sample2))
    print("T-Statistic:",round(t_stat, 4))
    print("P-Value:",round(p_value, 4))
    if p_value<alpha:
        print(f"Since p-value < {alpha}, we reject the null hypothesis.")
        print("There is a statistically significant difference between the two group means.")
    else:
        print(f"Since p-value >= {alpha}, we fail to reject the null hypothesis.")
        print("There is no statistically significant difference between the two group means.")
np.random.seed(42)
group1=np.random.normal(loc=100,scale=10,size=30)
group2=np.random.normal(loc=105,scale=10,size=30) 
independent_t_test(group1,group2,alpha=0.05)
```

```python
# 7. Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel
def paired_t_test_and_plot(before,after,alpha=0.05):
    t_stat,p_value=ttest_rel(before,after)
    print("Before Mean:",np.mean(before))
    print("After Mean:",np.mean(after))
    print("T-Statistic:",round(t_stat,4))
    print("P-Value:",round(p_value,4))
    if p_value<alpha:
        print(f"Since p-value < {alpha}, we reject the null hypothesis.")
        print("There is a statistically significant difference between before and after.")
    else:
        print(f"Since p-value >= {alpha}, we fail to reject the null hypothesis.")
        print("No statistically significant difference detected.")
    plt.figure(figsize=(10,6))
    for i in range(len(before)):
        plt.plot([1, 2],[before[i],after[i]],marker='o',color='gray',alpha=0.7)
    plt.xticks([1, 2],['Before','After'])
    plt.title('Paired Sample Comparison')
    plt.ylabel('Values')
    plt.grid(True)
    plt.show()
np.random.seed(0)
before_scores=np.random.normal(loc=70,scale=5,size=30)
after_scores=before_scores+np.random.normal(loc=3,scale=5,size=30) 
paired_t_test_and_plot(before_scores,after_scores)
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats
def confidence_interval(sample,confidence=0.95):
    n=len(sample)
    mean=np.mean(sample)
    std_err=stats.sem(sample) 
    margin=std_err*stats.t.ppf((1+confidence)/2,df=n-1) 
    lower=mean-margin
    upper=mean+margin
    return mean,(lower,upper)
np.random.seed(0)
sample_data=np.random.normal(loc=100,scale=15,size=30)
mean,ci=confidence_interval(sample_data)
print(f"Sample Mean = {mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")
```

```python
# 10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy import stats
def calculate_margin_of_error(sample,confidence=0.95):
    n=len(sample)
    std_err=stats.sem(sample)  
    t_critical=stats.t.ppf((1+confidence)/2,df=n-1)  
    margin_of_error=t_critical*std_err
    return margin_of_error
np.random.seed(42)
sample_data=np.random.normal(loc=75,scale=10,size=40)
moe=calculate_margin_of_error(sample_data,confidence=0.95)
print(f"Sample Size = {len(sample_data)}")
print(f"Sample Mean = {np.mean(sample_data):.2f}")
print(f"Margin of Error (95% Confidence) = ±{moe:.2f}")
```

```python
# 11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(p_disease,p_positive_given_disease,p_positive_given_no_disease):
    p_no_disease=1-p_disease
    p_positive=(p_positive_given_disease*p_disease)+(p_positive_given_no_disease*p_no_disease)
    p_disease_given_positive=(p_positive_given_disease*p_disease)/p_positive
    return p_disease_given_positive
p_disease=0.01
p_positive_given_disease=0.99
p_positive_given_no_disease=0.05
posterior=bayes_theorem(p_disease,p_positive_given_disease,p_positive_given_no_disease)
print(f"Probability of having the disease given a positive test result: {posterior:.4f}")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np
import pandas as pd
def calculate_expected_frequencies(observed):
    observed=np.array(observed)
    row_totals=observed.sum(axis=1,keepdims=True) 
    col_totals=observed.sum(axis=0,keepdims=True)  
    grand_total=observed.sum()
    expected=row_totals@col_totals/grand_total
    return expected
observed=[[30, 10],
            [20, 40]]
expected=calculate_expected_frequencies(observed)
df_expected=pd.DataFrame(expected,columns=['Product_A','Product_B'],index=['Male','Female'])
print("Expected Frequencies:")
print(df_expected)
```
